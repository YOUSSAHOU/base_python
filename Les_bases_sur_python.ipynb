Exercice 6: Le français sans voyelle

```python
def sans_voyelle (text):
    texte=""
    for c in text:
        for i in "aeiouy":
            if c==i:
                c=""
            else:
                c=c
        texte=texte + c
    return texte
        
```

```python
sans_voyelle("python")
```

```python
sans_voyelle("cahier")
```

Exercice 7: Des phrases sans voyelles autre que a


```python
def sans_voyelle(string):
    Phrase=""
    for c in string:
        for i in "eiouyEIOUY":
            if c==i:
                c=""
            else:
                c=c
        Phrase= Phrase+c
    return Phrase
```

```python
sans_voyelle("Il fait tres beau ce matin")
```

```python
sans_voyelle("Bonjour marlène. Comment allez-vous?")
```

Exercice 9: Dis moi ton index, je te dirai qui tu es


```python
def somme_pair_impair(a, n):
    somme_pair = sum(a[::2])
    somme_impair = sum(a[1::2])
    print("La somme des index pair", somme_pair)
    print("La somme des index impair", somme_impair)
    return ([somme_pair, somme_impair])
 
```

```python
Liste_nbres= ([10,20,30,40,50,60])
n=len(Liste_nbres)

somme_pair_impair(Liste_nbres, n)
```

Exercie 14 : Trop d'escale, c'est fatiguant

```python
ville_escale= input("la ville d'escale: ")
duree_escale= input("la durée d'escale: ")
duree_escale=float(duree_escale)
Dict[ville_escale]=duree_escale
print(Dict)
```
